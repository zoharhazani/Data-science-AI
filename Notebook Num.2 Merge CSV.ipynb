```python
import pandas as pd
```

## on the last notebook we download the pages on parts, in order not to be blocked.
## here we merge all the csv files that we create into one data frame.

```python
frames = []
```

```python
for i in range(1,39):
    df = pd.read_csv(f'./preply{i}.csv')
    frames.append(df)
```

```python
df_final = pd.concat(frames)
df_final
```

# Data Handling phase:
### after we got all of our data conceted into one chart, we will take care of the data.

## -first : we will remove all of the duplicated rows that contains the same people based on their names.
###### it can happen as a result of :
####  *section the information into different charts
#### *crawling on different days some features has changed (for example: number of people increased)
#### *changes on the web and craweling on the same time .
#### *also if the preply web contains pages with the same cards.

### all of those reasons affected our data so we will take care of it in order to get high data quality

```python
df_final[df_final.duplicated(['name'])]
```

### checking the number of duplicated rows: 

```python
df_final[df_final.duplicated()].shape
```

### and removing them:

```python
df_final.drop_duplicates()
```

#### We tried to find another duplicated rows based on the description(chances are low that two different people will have the same description) and teaches.

```python
df_final[df_final.duplicated(subset=['description','teaches'])].shape
```

```python
df_final.drop_duplicates(subset=['description','teaches'],keep='first',inplace= True)
df_final.drop(['Unnamed: 0'],axis=1,inplace=True)
df_final
```
